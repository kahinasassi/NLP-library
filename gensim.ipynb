## 3 create a bag of word

<b>bag of word</b> : word + id"index or token"+ frequence "occurance"

```python
from gensim.utils import simple_preprocess
from gensim import corpora

my_doc =["I'm student in master degree", "I'm working in NLP filed","the projest is HATE SPEECH in algerian dialect"]
#1 tokenize my_doc
tock_list=[simple_preprocess(doc) for doc in my_doc]
mydict=corpora.Dictionary()

#2 create a bag of word
mycorpus=[mydict.doc2bow(doc,allow_update=True) for doc in tock_list]
#3 afficher l'index de chque mots et son occurance
print(mycorpus)
print("\n")
print("\n")
word_count = [[(mydict[id],count) for id,count in line]for line in mycorpus]
print(word_count)
```

## 5 save and load a dictionary to/from disk

```python
from gensim.utils import simple_preprocess
from gensim import corpora
my_doc = ["Petra is in Jordan","Aman?","Aman?","Aman?","Aman?"]
#tokenize the doc
tok_list = [simple_preprocess(doc) for doc in my_doc]
#create a dictionary
mydict = corpora.Dictionary()
mycorp= [mydict.doc2bow(doc,allow_update=True) for doc in tok_list]
print(mycorp)
print("\n")

#save the dictionary and corpus
mydict.save('mydict.dict')#save dict
corpora.MmCorpus.serialize('bow_corpus.mm',mycorp)#save corpus

#load them back
loaded_dict = corpora.Dictionary.load('mydict.dict')
print(loaded_dict)

corpus=corpora.MmCorpus('bow_corpus.mm')
for line in corpus:
    print(line)
```

## 6 Create the Tf-idf matrix (corpus) in gensim

```python
from gensim import models
from gensim import corpora
from gensim.utils import simple_preprocess
import numpy as np

my_doc=["this is the first line", "this is the second line","this is the third line"]


tock_list=[simple_preprocess(doc) for doc in my_doc]
mydict=corpora.Dictionary()
# create a bag of word:
mycorpus=[mydict.doc2bow(doc,allow_update=True) for doc in tock_list]
word_count = [[(mydict[id],count) for id,count in line]for line in mycorpus]
print(word_count)
print("\n")

#create tf-idf weights:
tfidf = models.TfidfModel(corpus,smartirs='ntc')
#show the tf-idf weights:
for doc in tfidf[mycorpus]:
    print([[mydict[id],np.around(freq,decimals=2)] for id,freq in doc])
```

## 8 Doc2Vec: to find most similar doc using tags

```python
from gensim.models.doc2vec import Doc2Vec , TaggedDocument
from nltk.tokenize import word_tokenize
data=["I love machine learning","I really love deep learning","I love so much reinforsment learning"]
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i,_d in enumerate(data)]
print(tagged_data)

max_epoch = 100
vec_size = 20
alpha = 0.025

model =Doc2Vec(size=vec_size, alpha=alpha, min_alpha=0.00025, min_count=1, dm=1)
model.build_vocab(tagged_data)

for epoch in range(max_epoch):
    model.train(tagged_data  ,total_examples=model.corpus_count, epochs=model.iter)
    #decrease the learning rate
    model.alpha =-0.0002
    #fit the learning rate, no decay
    model.min_alpha = model.alpha
model.save("d2v.model")
print("Model is saved")
```

```python
model= Doc2Vec.load("d2v.model")
#to find the vector of a document which is not in the training data
test_data = word_tokenize("I live chatbots".lower())
v1 = model.infer_vector(test_data)
print("v1_infer:",v1)

#to find most similar doc using tags
similar_doc = model.docvecs.most_similar('1')
print("\n similar doc:",similar_doc)

#to find vect of doc in training data using tags or in other words,
#printing the vector of doc at index 1 in training data
print("\n model.docvecs[1]:",model.docvecs['1'])
```

## 9 visualize word embading

### Word2vec

```python
from gensim.models import Word2Vec # way to visualise  word embading
#define training data
sentenses =[['this','is','the','first','sentenses','for','word2vec'],
            ['this','is','the','first','sentenses'],
            ['yet','another','sentenses'],
            ['one','more','sentenses'],
            ['and','the','final','sentenses']]
#TRAIN THE MODEL
model= Word2Vec(sentenses,min_count=1)
#sammarize the loaded model
print("model:",model)
#sammarize the vocabulary
words= list(model.wv.vocab)
print("\n words:",words)
#access the vector for one word
print("\n model['sentenses']:",model['sentenses'])
#save model
model.save('model.bin')
#load model
new_model = Word2Vec.load('model.bin')

print("\n loaded model: " , new_model)
```

### visaulize

```python
from gensim.models import Word2Vec # way to visualise  word embading
from sklearn.decomposition import PCA
from matplotlib import pyplot
#define training data
sentenses =[['this','is','the','first','sentenses','for','word2vec'],['this','is','the','first','sentenses'],['yet','another','sentenses'],['one','more','sentenses'],['and','the','final','sentenses']]
#TRAIN THE MODEL
model= Word2Vec(sentenses,min_count=1)
#sammarize the loaded model
print("model:",model)
#fit a 2d PCA model to the vectors
x = model[model.wv.vocab ]
pca= PCA(n_components=2)
result = pca.fit_transform(x)

#create the scatter plot of the projection
pyplot.scatter(result[:,0],result[:,1])

words = list(model.wv.vocab)
for i,word in enumerate(words):
    pyplot.annotate(word,xy=(result[i,0],result[i,1]))
pyplot.show()
```

## 10 functions to process raw text

#### remove stop word

```python
from gensim.parsing.preprocessing import remove_stopwords
print(remove_stopwords('we are the only ones respnsible to create the life , we dream of. '))
```

#### separate the alphabet from nbr

```python
from gensim.parsing.preprocessing import split_alphanum
print(split_alphanum('we stady5 day per7days and 5hours in 24h. '))#le text doit etre en lower 
```

#### remove the s from noum that indicate plural

```python
from gensim.parsing.preprocessing import stem_text
print(stem_text('she has a lots of documments , because she regesters in lot of librarys '))
```

#### remove the white space : \n \t ''


```python
from gensim.parsing.preprocessing import strip_multiple_whitespaces
print(strip_multiple_whitespaces('she has a lots of documments '+"\n"+',            because she regesters in lot of librarys '+"\t"))
```

#### remove the alpha num caractere from text : #@%£$~ NOT A NUMBR

```python
from gensim.parsing.preprocessing import strip_non_alphanum
print(strip_non_alphanum('#she_has--a 4 lots of *documments* , £$because she regesters in lot of @librarys¨^¨ '))
```

#### remove the numerique value from text

```python
from gensim.parsing.preprocessing import strip_numeric
print(strip_numeric(' 123text is too lo5ng0000 LENGTH44')) 
```

#### remove ponctuation

```python
from gensim.parsing.preprocessing import strip_punctuation
print(strip_punctuation(' text is too long!!!!! , omg! what shall I do? ')) 
```

#### remove words having some length defined by us

```python
from gensim.parsing.preprocessing import strip_short
print(strip_short(' text is too long ! , omg! what shall I do ? ',minsize=3)) # all word having length<3
```

```python
from gensim.parsing.preprocessing import strip_tags
print(strip_tags(' <li>text is too long ! , <b>omg!</b> what shall I do ? </li>')) 
```
